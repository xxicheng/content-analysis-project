```python
large_df = pd.read_table('../data/preprocessed inequality news.txt' ,header=0, encoding='gb2312', sep=',') 
```

```python
large_df.dropna(inplace=True)
```

```python
def get_topic_file(file_name):
    file_path = '../data/inequality_topic/'
    file = file_path + file_name
    with open(file, 'r') as read_obj:
        # pass the file object to reader() to get the reader object
        csv_reader = reader(read_obj)
        # Pass reader object to list() to get a list of lists
        cluster = list(csv_reader)
        
    ids = []
    for ele in cluster: # list_of_rows[1] are id
        ele = list(filter(lambda x: x!= '', ele))
        ele = str(ele)
        for i, text in enumerate(large_df["reduced_tokens"]):
            #print(ele, type(ele))
            #print(text, type(text))
            if ele == text :
                ids.append(large_df['id'][i])
    
    print(len(ids))
    outcome_df = large_df[large_df['id'].isin(ids)]
    
    return outcome_df
```

```python
outcome_df = get_topic_file('topic0.csv')
outcome_df.to_csv('../data/inequality_topic/topic0_text.csv' )
```

```python
for i in range(9):
    outcome_df = get_topic_file('topic'+ str(i) +'.csv')
    outcome_df.to_csv('../data/inequality_topic/topic'+ str(i) +'_text.csv' )
```
